```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
hotel = pd.read_csv('../processed/hotel_dataset_modificado.csv')
hotel
```

```python
X = hotel.drop(['booking_status'], axis= 1)
y = hotel['booking_status']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
train = pd.concat([X_train, y_train], axis= 1)
```

```python
nombre_archivo = 'train.csv'
```

```python
train.to_csv(nombre_archivo, index=False)
```

```python
test = pd.concat([X_test, y_test], axis= 1)
```

```python
nombre_archivo_test = 'test.csv'
```

```python
test.to_csv(nombre_archivo_test, index=False)
```
